```python
import json
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
with open('keys.json') as fi:
    credentials = json.load(fi)
```

```python
state="me"
endpoint = f"https://developer.nps.gov/api/v1/parks?stateCode={state}"

HEADERS = {"X-Api-Key":credentials["api_key"]}

req = requests.get(endpoint,headers=HEADERS)
res=req.json()
res
```

```python
park_info= pd.json_normalize(res['data'])
```

```python
park_info.head()
```

```python
len(park_info['activities'][0])
```
