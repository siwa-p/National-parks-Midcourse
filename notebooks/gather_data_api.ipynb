```python
import json
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
with open('../keys.json') as fi:
    credentials = json.load(fi)
```

```python
states = pd.read_csv('../data/states.csv')
```

```python
states['Abbreviation']
```

```python
dfs=[]
res=[]
for state in states['Abbreviation']:
    endpoint = f"https://developer.nps.gov/api/v1/parks?stateCode={state}"

    HEADERS = {"X-Api-Key":credentials["api_key"]}

    req = requests.get(endpoint,headers=HEADERS)
    res.append(req.json())
for i in range(len(res)):
    dfs.append(pd.json_normalize(res[i]['data']))

park_info = pd.concat(dfs, ignore_index=True)
park_info
```

```python
park_info.to_csv('../data/park_info.csv')
```
